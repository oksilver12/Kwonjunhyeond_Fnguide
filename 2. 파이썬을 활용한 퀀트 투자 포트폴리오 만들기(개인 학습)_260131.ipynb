### 8.2 크롤링 예제

#### 8.2.2 금융 속보 크롤링

```python
# @title 기사 제목 크롤링
# ① 웹 크롤링을 위한 라이브러리 임포트
# 이 섹션에서는 웹 페이지의 HTML 내용을 가져오고 분석하는 데 필요한 라이브러리들을 불러옵니다.
import requests as rq # requests 모듈을 rq라는 별칭으로 임포트 // 웹 페이지에 HTTP 요청을 보내는 라이브러리
from bs4 import BeautifulSoup # bs4 모듈에서 BeautifulSoup 클래스 임포트 // 가져온 HTML 내용을 파싱(분석)하여 필요한 데이터를 추출하는 라이브러리

# ① 웹 페이지 데이터 가져오기 및 파싱
# 이 섹션에서는 특정 금융 뉴스 웹 페이지에 접속하여 HTML 데이터를 가져오고 BeautifulSoup을 사용하여 파싱합니다.
url = 'https://finance.naver.com/news/news_list.nhn?mode=LSS2D&section_id=101&section_id2=258' # url 변수에 = 'https://finance.naver.com/...' 문자열 할당 // 네이버 금융 뉴스 페이지 URL 정의
data = rq.get(url) # data 변수에 = rq(requests 모듈)에서 . get 함수로 (url에) 요청 // 지정된 URL에서 웹 페이지 데이터 가져오기
html = BeautifulSoup(data.content, 'html.parser') # html 변수에 = BeautifulSoup으로 (data.content를, 'html.parser'로) 파싱 // 가져온 HTML 데이터를 BeautifulSoup 객체로 파싱
html_select = html.select('dl > dd.articleSubject > a') # html_select 변수에 = html 객체에서 . select 함수로 ('dl > dd.articleSubject > a') CSS 선택자 적용 // 특정 CSS 선택자를 사용하여 뉴스 기사 제목 요소들 추출

# ② 추출된 요소 미리보기
# 추출된 뉴스 기사 제목 요소들 중 일부를 출력하여 구조를 확인합니다.
html_select[0:3] # html_select 리스트의 [0번째부터 3번째 전까지] 요소 출력 // 첫 3개의 뉴스 기사 제목 요소 출력

# ① 첫 번째 뉴스 기사의 제목 추출
# 이 섹션에서는 이전에 추출한 뉴스 기사 요소들 중 첫 번째 기사의 'title' 속성 값을 추출하여 출력합니다.
html_select[0]['title'] # html_select 리스트의 [0번째 즉, 첫 번째] 요소에서 ['title'] 속성 값 추출 // 첫 번째 뉴스 기사의 제목 추출

# ① 모든 뉴스 기사의 제목 추출
# 이 섹션에서는 이전에 추출한 모든 뉴스 기사 요소들의 'title' 속성 값을 추출하여 리스트 형태로 반환합니다.
[i['title'] for i in html_select] # i의 ['title'] 속성 값을 추출 (for i는 html_select 안에 있는) // html_select 리스트의 각 요소에서 'title' 속성을 추출하여 새 리스트 생성
```

#### 8.2.4 기업 공시 채널에서 오늘의 공시 불러오기

```python
# @title 라이브러리 불러오기

import requests as rq # 웹 페이지에 요청을 보내는 라이브러리
from bs4 import BeautifulSoup # HTML 및 XML 파일을 파싱하는 라이브러리
import pandas as pd # 데이터 분석 및 조작을 위한 라이브러리
```

```python
# @title 공시 데이터 불러오기

# ① URL과 쿼리를 입력한다. 쿼리는 딕셔너리 형태로 입력하며, Form Data와 동일하게 입력해 준다.
# 쿼리 중 marketType과 같이 값이 없는 항목은 입력하지 않아도 된다.
url = 'https://kind.krx.co.kr/disclosure/todaydisclosure.do' # 요청을 보낼 웹사이트 URL
payload = {
'method': 'searchTodayDisclosureSub', # 호출할 메소드 이름
'currentPageSize': '15', # 한 페이지당 보여줄 항목 수
'pageIndex': '1', # 현재 페이지 인덱스
'orderMode': '0', # 정렬 모드
'orderStat': 'D', # 정렬 상태 (예: 날짜 기준)
'forward': 'todaydisclosure_sub', # 전달할 페이지 정보
'chose': 'S', # 선택된 값 (예: 'S'는 특정 유형의 공시)
'todayFlag': 'N', # 오늘 공시 여부 ('N'은 오늘이 아님)
'selDate': '2025-02-11' # 조회할 날짜
}

# ② POST() 함수를 통해 해당 URL에 원하는 쿼리를 요청한다.
data = rq.post(url, data=payload) # 지정된 URL과 페이로드로 POST 요청 전송

# ③ BeautifulSoup() 함수를 통해 파싱한다.
html = BeautifulSoup(data.content, 'html.parser') # 응답받은 HTML 내용을 BeautifulSoup 객체로 파싱

print(html)
```

```python
# @title 공시데이터 확인하기

# ① prettify() 함수를 이용해 BeautifulSoup에서 파싱한 파서 트리를 유니코드 형태로 다시 돌려준다.
html_unicode = html.prettify() # BeautifulSoup 객체의 HTML 내용을 들여쓰기하여 보기 좋게 만듭니다.

# ② read_html() 함수를 통해 표를 읽어 온다.
tbl = pd.read_html(html.prettify()) # prettify()된 HTML에서 테이블을 찾아 리스트 형태로 반환하고, 첫 번째 테이블을 선택합니다.

tbl[0].head() # 파싱된 첫 번째 테이블(DataFrame)의 상위 5개 행을 출력하여 데이터 구조를 확인합니다.
```

### 10.1 최근 영업일 기준 데이터 받기

```python
# @title 날짜 추출하기
# ① 라이브러리 임포트
# 이 섹션에서는 웹 크롤링에 필요한 라이브러리를 가져옵니다.
import requests as rq # requests 라이브러리를 rq라는 별칭으로 임포트 // 웹 페이지에 HTTP 요청을 보내는 라이브러리
from bs4 import BeautifulSoup # bs4 모듈에서 BeautifulSoup 클래스를 임포트 // HTML 내용을 파싱(분석)하여 필요한 데이터를 추출하는 라이브러리

# ② 웹 페이지 데이터 가져오기 및 파싱
# 이 섹션에서는 네이버 금융에서 최신 영업일 날짜 정보를 포함하는 웹 페이지를 가져와 파싱합니다.
url = 'https://finance.naver.com/sise/sise_deposit.nhn' # url 변수에 = 'https://finance.naver.com/sise/sise_deposit.nhn' 문자열을 할당 // 네이버 금융 시세 및 예탁금 현황 페이지 URL 정의
data = rq.get(url) # data 변수에 = rq(requests 모듈)에서 .get 함수로 (url에) 요청 // 지정된 URL에서 웹 페이지 데이터 가져오기
data_html = BeautifulSoup(data.content) # data_html 변수에 = BeautifulSoup으로 (data.content를) 파싱 // 가져온 HTML 데이터를 BeautifulSoup 객체로 파싱

# ③ 날짜 정보 추출 및 출력
# 이 섹션에서는 파싱된 HTML에서 특정 CSS 선택자를 이용하여 영업일 날짜를 추출하고 출력합니다.
parse_day = data_html.select_one(
    'div.subtop_sise_graph2>ul.subtop_chart_note>li>span.tah').text # parse_day 변수에 = data_html 객체에서 .select_one 함수로 ('div.subtop_sise_graph2>ul.subtop_chart_note>li>span.tah') CSS 선택자를 적용하여 찾은 요소의 .text 값 할당 // 특정 CSS 선택자를 사용하여 영업일 날짜 텍스트 추출

print(parse_day) # print 함수로 (parse_day를) 출력 // 추출된 영업일 날짜 출력
```
